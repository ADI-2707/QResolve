# Date Preparation for Time-Series Analysis

Convert the purchase date into datetime format for time-series analysis.

```python
df["Date of Purchase"] = pd.to_datetime(df["Date of Purchase"])

df["Date of Purchase"].head()
```

# Monthly Ticket Trend

Visualize how purchases are distributed over time.

```python
monthly_tickets = (
    df.groupby(df["Date of Purchase"].dt.to_period("M"))
    .size()
)

monthly_tickets.index = monthly_tickets.index.astype(str)

plt.figure(figsize=(12, 5))

plt.plot(monthly_tickets.index, monthly_tickets.values, marker="o")

plt.title("Monthly Purchase Trend")
plt.xlabel("Month")
plt.ylabel("Number of Purchases")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# Priority vs Customer Satisfaction

Analyze whether ticket priority affects customer satisfaction ratings.

```python
priority_rating = (
    df.groupby("Ticket Priority")["Customer Satisfaction Rating"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(6,4))

priority_rating.plot(kind="bar")

plt.title("Average Satisfaction by Ticket Priority")
plt.xlabel("Ticket Priority")
plt.ylabel("Average Satisfaction Rating")

plt.show()
```

# First Response Time Analysis
Analyze the most common first response times for customer support tickets.

```python
response_time = (
    df["First Response Time"]
      .value_counts()
      .head(10)
)

plt.figure(figsize=(10,5))
response_time.plot(kind="bar")
plt.title("Most Common First Response Times")
plt.xlabel("First Response Time")
plt.ylabel("Number of Tickets")
plt.xticks(rotation=45)
plt.show()
```

# Resolution Time Analysis

Analyze how long tickets take to be resolved.

```python
resolution_time = (
    df["Time to Resolution"]
      .value_counts()
      .head(10)
)

plt.figure(figsize=(10,5))
resolution_time.plot(kind="bar")
plt.title("Most Common Resolution Times")
plt.xlabel("Resolution Time")
plt.ylabel("Number of Tickets")
plt.xticks(rotation=45)
plt.show()
```

# Product-wise Customer Satisfaction

Analyze which purchased products receive higher customer satisfaction ratings.

```python
product_rating = (
    df.groupby("Product Purchased")["Customer Satisfaction Rating"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))

product_rating.plot(kind="bar")

plt.title("Top 10 Products by Average Customer Satisfaction")
plt.xlabel("Product")
plt.ylabel("Average Satisfaction Rating")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
