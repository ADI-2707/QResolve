# Ticket Status Distribution

Visualize the distribution of ticket statuses to understand the current support workload.

```python
status_counts = df["Ticket Status"].value_counts()

plt.figure(figsize=(6, 4))
status_counts.plot(kind="bar")

plt.title("Ticket Status Distribution")
plt.xlabel("Ticket Status")
plt.ylabel("Number of Tickets")

plt.show()
```

# Ticket Type Distribution

Visualize the frequency of different ticket types.

```python
ticket_type_counts = df["Ticket Type"].value_counts()

plt.figure(figsize=(8, 5))
ticket_type_counts.plot(kind="bar")

plt.title("Ticket Type Distribution")
plt.xlabel("Ticket Type")
plt.ylabel("Number of Tickets")

plt.show()
```

# Ticket Priority Distribution

Analyze the distribution of ticket priorities.

```python
priority_counts = df["Ticket Priority"].value_counts()

plt.figure(figsize=(6,4))
priority_counts.plot(kind="bar")

plt.title("Ticket Priority Distribution")
plt.xlabel("Priority")
plt.ylabel("Number of Tickets")

plt.show()
```

# Customer Gender Distribution

Inspect the distribution of customers by gender.

```python
gender_counts = df["Customer Gender"].value_counts()

plt.figure(figsize=(6,4))
gender_counts.plot(kind="bar")

plt.title("Customer Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")

plt.show()
```

# Top Products Purchased

Display the top 10 most frequently purchased products.

```python
top_products = df["Product Purchased"].value_counts().head(10)

plt.figure(figsize=(10,5))
top_products.plot(kind="bar")

plt.title("Top 10 Purchased Products")
plt.xlabel("Product")
plt.ylabel("Number of Purchases")

plt.xticks(rotation=45)

plt.show()
```

# Customer Age Distribution

Visualize the distribution of customer ages.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Customer Age"], bins=15)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")

plt.show()
```

# Customer Satisfaction Rating

Analyze the distribution of customer satisfaction ratings.

```python
rating_counts = df["Customer Satisfaction Rating"].value_counts().sort_index()

plt.figure(figsize=(6,4))
rating_counts.plot(kind="bar")

plt.title("Customer Satisfaction Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Number of Customers")

plt.show()
```
