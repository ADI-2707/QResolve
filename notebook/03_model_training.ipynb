# Model Training

In this notebook, we train and evaluate multiple machine learning regression models using the processed customer support dataset.

## Objectives

- Load processed training and testing datasets
- Train multiple regression models
- Evaluate model performance
- Compare models
- Save the best performing model

```python
import pandas as pd
import joblib
```

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import joblib

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import (
    RandomForestRegressor,
    GradientBoostingRegressor
)

from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
```

## Load Processed Dataset

Load the processed training and testing datasets generated during the feature engineering stage.

```python
import joblib

data = joblib.load("../processed/train_test_data.pkl")

X_train = data["X_train"]
X_test = data["X_test"]
y_train = data["y_train"]
y_test = data["y_test"]
```

## Prepare Target Variables

Convert the target variables from DataFrames to Pandas Series for compatibility with scikit-learn estimators.

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

## Verify Dataset Shapes

Inspect the dimensions of the training and testing datasets.

```python
print("X_train:", X_train.shape)
print("X_test :", X_test.shape)
```

# Import Classification Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix,
)
```

# Inspect Target Classes

```python
print("Target data type:", y_train.dtype)
print()

print("Unique classes:")
print(sorted(y_train.unique()))

print()

print("Class distribution:")
print(y_train.value_counts())
```

# Logistic Regression Classifier

```python
logistic_model = LogisticRegression(
    max_iter=1000,
    random_state=42
)

logistic_model.fit(X_train, y_train)
```

```python
logistic_pred = logistic_model.predict(X_test)

accuracy = accuracy_score(y_test, logistic_pred)

print(f"Accuracy: {accuracy:.4f}")

print("\nClassification Report")
print(classification_report(y_test, logistic_pred))
```

# Decision Tree Classifier

```python
dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(X_train, y_train)

dt_pred = dt_model.predict(X_test)

dt_accuracy = accuracy_score(y_test, dt_pred)

print(f"Decision Tree Accuracy: {dt_accuracy:.4f}")

print("\nClassification Report")
print(classification_report(y_test, dt_pred))
```

# Random Forest Classifier

```python
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Random Forest Accuracy: {rf_accuracy:.4f}")

print("\nClassification Report")
print(classification_report(y_test, rf_pred))
```

# Gradient Boosting Classifier

```python
gb_model = GradientBoostingClassifier(random_state=42)

gb_model.fit(X_train, y_train)

gb_pred = gb_model.predict(X_test)

gb_accuracy = accuracy_score(y_test, gb_pred)

print(f"Gradient Boosting Accuracy: {gb_accuracy:.4f}")

print("\nClassification Report")
print(classification_report(y_test, gb_pred))
```

# Model Comparison

```python
results = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "Decision Tree",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Accuracy": [
        accuracy,
        dt_accuracy,
        rf_accuracy,
        gb_accuracy
    ]
})

results = results.sort_values(by="Accuracy", ascending=False)

results
```
